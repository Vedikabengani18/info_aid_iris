```python
#create an class
class LabelEncoder(object):
  def __init__(self):
    self.classes_=None#classes will store the list in1d->display the index in form of true and false and give it in form of 1d array
  def fit(self,labels:list)->None:#first fit should be call
    self.classes_=sorted(set(labels))#if 2 underscore is there then it it a private variable and
    #if at end then it is just for notation only and in python there is no private variable you can access itt ousie the variable;
  def fit_transform(self,labels)->list:#transfor decode the code and provide the list
    self.fit(labels)
    return[self.classes_.index(label) for label in labels]
```
